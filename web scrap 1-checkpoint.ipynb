```python
import time
from selenium import webdriver
from bs4 import BeautifulSoup
import pandas as pd

# URL of the page to scrape
url = "https://housing.com/paying-guests/search-P4bimjmco2m9afw0mS1?listingId=11381669%23roomsOffered"

# Initialize a webdriver object and load the page
driver = webdriver.Chrome()
driver.get(url)

# Wait for the page to fully load (you can adjust the sleep time as needed)
time.sleep(5)

# Get the HTML content of the page after it has loaded dynamically
html = driver.execute_script("return document.documentElement.outerHTML")

# Close the webdriver
driver.quit()

# Parse the HTML content using Beautiful Soup
soup = BeautifulSoup(html, 'html.parser')

# Find all the Paying Guest items on the page
pg_items = soup.find_all('section', {'class': 'css-10nd30h'})

print(pg_items)
```

```python
# Create an empty list to store the data
data = []

# Loop through each Paying Guest item and extract the name and location
for pg_item in pg_items:
    name = pg_item.find('a', {'data-q': 'title'}).text
    location = pg_item.find('a', {'class': 'link last-link _18uq1994 _ot7i1994'}).text
    
    price_label = pg_item.find('label', {'class': 'css-10oy6g6'})
    price_range = price_label.text.strip()
    
    container = pg_item.find('div', {'class': 'css-qjg9zr'})
    girls = container.find('div', {'class': 'css-1fnf2a6'})
    boys = container.find('div', {'class': 'css-jfno6a'})

    if girls is not None:
        girls = girls.text
    else:
        girls = "Not available"

    if boys is not None:
        boys = boys.text
    else:
        boys = "Not available"
        
    room_types = pg_item.find_all('div', {'class': 'css-1i4xnr1'}) + pg_item.find_all('div', {'class': 'css-16k5c9z'})

    room_type_list = []

    for room_type in room_types:
        room_type_text = room_type.find('div', {'class': 'css-1l0lfmx'}).text.strip()
        if room_type_text:
            sharing_type_text = room_type.find('div', {'class': 'css-158dewz'}).text.strip()
            room_type = f"{room_type_text} {sharing_type_text}"
            if room_type not in room_type_list:
                room_type_list.append(room_type)
        else:
            room_type_list.append("Not available")
    # Append the data to the list
    data.append([name, location,price_range,girls,boys,room_type_list])

# Create a pandas dataframe from the list of data
df = pd.DataFrame(data, columns=['Name', 'Location','Price_Range','Girls','Boys','room_type_list'])

df
```
